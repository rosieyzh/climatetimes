```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_links(result_link):
    links=[]
    r = requests.get(result_link)
    soup = BeautifulSoup(r.text, 'html.parser')  
    article_results = soup.find("div", {"class":"col-8 nocontent"})
    for ana in article_results.findAll("a"):
        if ana.parent.name == "article" and 'video' not in ana['href'] and 'picture' not in ana['href']:
            links.append(ana['href'])
    return links

def get_text(article_link):
    article=""
    title=''
    date=''
    author=''
    r = requests.get(article_link)
    soup = BeautifulSoup(r.text, 'html.parser')
    #get title, date, author
    if soup.find('h1',{'class':'content__title'}):
        title=soup.find('h1',{'class':'content__title'}).text
    else:
        title=soup.find('title').text
    if soup.find('p',{'class':'content__meta-byline'}):
        author=soup.find('p',{'class':'content__meta-byline'}).text.replace("By ","")
    else:
        if soup.find('span', {'class':'source'}):
            author=soup.find('span', {'class':'source'}).text
    if soup.find('p',{'class':'content__meta-timestamp'}):
        date=soup.find('p',{'class':'content__meta-timestamp'}).text
    else:
        if soup.find('span',{'class':'time'}):
            date=soup.find('span',{'class':'time'}).text
    
    col_8 = soup.find(attrs={"data-page":"1"})
    if col_8:
        for p in col_8.findAll("p"):
            excerpt = p.text.replace(u'\xa0', u' ')
            article = article + "¶" + excerpt
    return article,title,date,author
```

```python
print(get_text('https://www.cbsnews.com/news/greta-thunberg-climate-activist-set-to-arrive-in-new-york-city/'))
```

```python
cbs_articles=[]
cbs_titles=[]
cbs_authors=[]
cbs_dates=[]

for i in range(1, 31):
    print(i)
    links = get_links("https://www.cbsnews.com/feature/climate-change/{}/".format(i))
    for link in links:
        print(link)
        article, title, date, author= get_text(link)
        cbs_articles.append(article)
        cbs_authors.append(author)
        cbs_dates.append(date)
        cbs_titles.append(title)
        
```
