```python
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
from selenium.webdriver.common.action_chains import ActionChains

# I used Firefox; you can use whichever browser you like.
browser = webdriver.Chrome()


# Tell Selenium to get the URL you're interested in.
browser.get("https://www.foxnews.com/category/us/environment/climate-change")

'''
# Close Alert Page
alert = browser.find_element_by_css_selector('.alert-banner .close')
alert.click()
time.sleep(3)
'''



while True:
    try:
        button = browser.find_element_by_link_text('Show More')
        ActionChains(browser).move_to_element(button).perform()
        button.click()
        time.sleep(4)
    except Exception as e:
        print (e)
        break

# Now that the page is fully scrolled, grab the source code.
source_data = browser.page_source

# Throw your source into BeautifulSoup and start parsing!
bs_data = BeautifulSoup(source_data, 'html.parser')
```

```python
article_list = bs_data.find('div', {'class':'article-list'})
links=[]
fox_titles=[]

# find all a tags
for a in article_list.findAll('a'):
    if a.parent.name == 'h4'and 'video.foxnews' not in a['href']:
        links.append('https://www.foxnews.com'+ a['href'])
        fox_titles.append(a.text)
```

```python
fox_titles[55:60]
```

```python
import requests
import pandas as pd

fox_articles=[]
fox_dates=[]
fox_authors=[]

for i in range(len(links)):
    print(links[i])
    article=''
    r = requests.get(links[i])
    soup = BeautifulSoup(r.text, 'html.parser')
    
    # get dates and author
    author_meta = soup.find('div', {'class':'article-meta-lower'})
    date_meta = soup.find('div', {'class':'article-meta-upper'})
    if not author_meta or not date_meta:
        print(i)
        continue
    fox_authors.append(author_meta.find('div',{'class':'author-byline'}).text.replace('By ',''))
    fox_dates.append(date_meta.find('time').text)
    
    # scrap article text
    article_container = soup.find('div', {'class':'article-body'})
    if article_container:
        featured = article_container.find('div', {'class':'featured'})
        if featured:
            featured.decompose()
        paragraphs = article_container.findAll('p')
        if paragraphs:
            for p in paragraphs:
                if "CLICK HERE" not in p.text:
                    article = article + "¶" + p.text.replace('\xa0', ' ')
            fox_articles.append(article)
```

```python
df = pd.DataFrame(fox_articles, columns=['article'])
df['title']=fox_titles
df['author']=fox_authors
df['date']=fox_dates
df['denial?'] = [1]*len(df.index)
cols=['title','author','date','article','denial?']
df=df[cols]
df = df.drop_duplicates()
df['author'] = df['author'].str.strip().replace('By', '', regex=False)
df.to_csv('../../data/right/data_fox_articles.csv', index=False)
```
