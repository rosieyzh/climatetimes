```python
"""Checks the length of each article in a CSV file and later removes it from the CSV file
"""
import pandas as pd
import numpy as np
```

```python
WEB_NAME = 'fox_articles'
df = pd.read_csv('../data/right/data_{}.csv'.format(WEB_NAME))
df.head()
```

```python
# len(df['article'].iloc[1])
df['length'] = [0] * len(df.index)
df.head()
```

```python
for i in df.index:
    df['length'].iat[i] = len(df['article'].iloc[i])

df.head()
```

```python
sort_by_length = df.sort_values('length')
sort_by_length.head()
```

```python
# MIN_NUM_CHARS = 227
MIN_NUM_CHARS = 1000

#Deleting cells of length 1
no_small_articles = sort_by_length[sort_by_length.length >= MIN_NUM_CHARS].head()

print(no_small_articles['article'].iloc[0])
no_small_articles.head()
```

```python
# no_small_articles = no_small_articles.drop('length', axis=1)
no_small_articles.to_csv('../data/filtered articles/{}_filtered.csv'.format(WEB_NAME), index=False)
no_small_articles.head()
```
