```python
"""Checks the length of each article in a CSV file and later removes it from the CSV file
"""
import pandas as pd
import numpy as np
```

```python
FILE_NAME = 'all_right'
df = pd.read_csv('../../data/{}.csv'.format(FILE_NAME))
df.head()
```

```python
#Removing index numbers column
df = df.drop('Unnamed: 0.1',axis=1)
df.head()
```

```python
# len(df['article'].iloc[1])
df['length'] = [0] * len(df.index)
df.head()
```

```python
#Get rid of nan content rows
df2 = df[df['content'].notnull()]

#Get the length of each article
for i in range(len(df2)):
    df2['length'].iat[i] = len(df2['content'].iloc[i])

df2.head()
```

```python
# df['content'].iloc[i]
# np.nan
```

```python
sort_by_length = df2.sort_values('length')
sort_by_length.head()
```

```python
MIN_NUM_CHARS = 1000

#Drops rows with small articles
no_small_articles = sort_by_length[sort_by_length.length >= MIN_NUM_CHARS]

print(no_small_articles['content'].iloc[0])
len(no_small_articles)
```

```python
# no_small_articles = no_small_articles.drop('length', axis=1)
no_small_articles.to_csv('../../data/{}_filtered.csv'.format(FILE_NAME), index=False)
no_small_articles.head()
```
