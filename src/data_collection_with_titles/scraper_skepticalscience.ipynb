```python
#links to climate misinformer profile page
links=[
    'https://skepticalscience.com/skeptic_Bob_Carter.htm',
    'https://skepticalscience.com/skeptic_John_Christy.htm',
    'https://skepticalscience.com/skeptic_Judith_Curry.htm',
    'https://skepticalscience.com/skeptic_William_Happer.htm',
    'https://skepticalscience.com/skeptic_Richard_Lindzen.htm',
    'https://skepticalscience.com/skeptic_Patrick_Michaels.htm',
    'https://skepticalscience.com/skeptic_Lubos_Motl.htm',
    'https://skepticalscience.com/skeptic_Richard_Muller.htm',
    'https://skepticalscience.com/skeptic_Kary_Mullis.htm',
    'https://skepticalscience.com/skeptic_Roger_Pielke_Sr.htm',
    'https://skepticalscience.com/skeptic_Ian_Plimer.htm',
    'https://skepticalscience.com/skeptic_Murry_Salby.htm',
    'https://skepticalscience.com/skeptic_Fred_Singer.htm',
    'https://skepticalscience.com/skeptic_Roy_Spencer.htm',
    'https://skepticalscience.com/skeptic_Tony_Abbott.htm',
    'https://skepticalscience.com/skeptic_Michele_Bachmann.htm',
    'https://skepticalscience.com/skeptic_James_Inhofe.htm',
    'https://skepticalscience.com/skeptic_Dennis_Jensen.htm',
    'https://skepticalscience.com/skeptic_Nigel_Lawson.htm',
    'https://skepticalscience.com/skeptic_Sarah_Palin.htm',
    'https://skepticalscience.com/skeptic_Ron_Paul.htm',
    'https://skepticalscience.com/skeptic_Tim_Pawlenty.htm',
    'https://skepticalscience.com/skeptic_Rick_Perry.htm',
    'https://skepticalscience.com/skeptic_Malcolm_Roberts.htm',
    'https://skepticalscience.com/skeptic_Dana_Rohrabacher.htm',
    'https://skepticalscience.com/skeptic_Marco_Rubio.htm',
    'https://skepticalscience.com/skeptic_David_Archibald.htm',
    'https://skepticalscience.com/skeptic_David_Evans.htm',
    'https://skepticalscience.com/skeptic_John_McLean.htm',
    'https://skepticalscience.com/skeptic_Monckton_Myths.htm',
    'https://skepticalscience.com/skeptic_David_Montgomery.htm',
    'https://skepticalscience.com/skeptic_Lawrence_Solomon.htm',
    'https://skepticalscience.com/skeptic_Joe_Bastardi.htm',
    'https://skepticalscience.com/skeptic_William_Kininmonth.htm',
    'https://skepticalscience.com/skeptic_Joe_Kernen.htm',
    'https://skepticalscience.com/skeptic_Matt_Ridley.htm',
    'https://skepticalscience.com/skeptic_David_Rose.htm'
]

#regex search to extract quote
b = re.compile(r'\"(.*)\"\d')

#list to store quotes
denial_quotes=[]
```

```python
for link in links:
    r = requests.get(link, verify=False)
    soup = BeautifulSoup(r.text, 'html.parser')
    tablerows = soup.findAll("tr", {"class":"footnote"})
    for entry in tablerows:
        climate_myth_cell = entry.findAll("td")[0].text
        stripped_quote=re.search(b, climate_myth_cell)
        if stripped_quote:
            quote=stripped_quote.group(1)
        denial_quotes.append(quote)
```

```python
#create Dataframe to store quotes
df = pd.DataFrame(denial_quotes, columns=['text']).drop_duplicates()
df['denial?'] = [1]*len(df.index)
df
```

```python
#save df to csv
df.to_csv('../data/data_skepticalscience.csv', index=False)
```
