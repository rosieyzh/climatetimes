```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
r = requests.get("https://www.desmogblog.com/global-warming-denier-database")
soup = BeautifulSoup(r.text, 'html.parser')
h2_container = soup.findAll('h2')

#for extra text cleaning
b = re.compile(r'(.*)\[\d*\]')
quotes=[]
```

```python
for a_list in soup.findAll('a'):
    if a_list.parent.name=='h2':
        profile=requests.get(a_list['href'])
        soup_profile = BeautifulSoup(profile.text, 'html.parser')
        for quote in soup_profile.findAll('blockquote'):
            stripped_quote=re.search(b, quote.text)
            if stripped_quote:
                quotes.append(re.sub('”', '', re.sub('“', '', stripped_quote.group(1))).replace(u'\xa0', u' '))
```

```python
len(quotes)
```

```python
df = pd.DataFrame(quotes, columns=['text']).drop_duplicates()
df['denial?'] = [1]*len(df.index)
df
```

```python
#save df to csv
df.to_csv('../data/data_desmog.csv', index=False)
```
